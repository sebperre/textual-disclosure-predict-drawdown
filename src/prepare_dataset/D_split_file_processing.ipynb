```python
import pandas as pd
from pathlib import Path
```

```python
save_path = Path("../data/ret_sample_processed.csv")

if not save_path.exists():
    wrote_header = False
    keep_cols = ["id", "gvkey", "excntry", "stock_ret", "prc", "ret_eom", "date"]
    for chunk in pd.read_csv("../data/ret_sample.csv", usecols=keep_cols, chunksize=1000000):
        chunk = chunk[chunk["excntry"] == "USA"]

        chunk.to_csv(
                save_path,
                index=False,
                mode="a",
                header=not wrote_header,
            )
        wrote_header = True
    print(f"Saved processed file to {save_path}.")
else:
    print(f"{save_path} already exists, not overwriting.")
```
