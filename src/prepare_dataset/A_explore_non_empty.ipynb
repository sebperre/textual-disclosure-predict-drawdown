In this notebook, I looked inside the text data.
We find that 2005 has too many missing risk factor discussions, 
but 2006-2024 are good enough ~80%+ non-empty.
 

```python
import pickle
import pandas as pd
from pathlib import Path

from constants import *
```

```python
# Look inside text_data/2005.zip
file_path = TEXT_DATA_FOLDER + "text_us_2023.pkl"
with open(file_path, "rb") as f:
    data = pickle.load(f)
    text_data_df = data
```

```python
text_data_df = text_data_df.reset_index(inplace=False)
```

```python
# Loop each year
#    check sum of non empty rf & mgmt
#    check sum of empty rf & mgmt

results_df = pd.DataFrame(columns=["year", "non_empty_rf", "empty_rf", "non_empty_mgmt", "empty_mgmt"])

for year in range(2005, 2025):
    file_path = TEXT_DATA_FOLDER + f"text_us_{year}.pkl"
    with open(file_path, "rb") as f:
        data = pickle.load(f)
        text_data_df = data
    text_data_df = text_data_df.reset_index(inplace=False)
    non_empty_rf = text_data_df[text_data_df["rf"] != '']
    empty_rf = text_data_df[text_data_df["rf"] == '']
    non_empty_mgmt = text_data_df[text_data_df["mgmt"] != '']
    empty_mgmt = text_data_df[text_data_df["mgmt"] == '']
    results_df = pd.concat([results_df, pd.DataFrame({
        "year": [year],
        "non_empty_rf": [len(non_empty_rf)],
        "empty_rf": [len(empty_rf)],
        "non_empty_mgmt": [len(non_empty_mgmt)],
        "empty_mgmt": [len(empty_mgmt)]
    })], ignore_index=True)
```

```python
text_data_df
```

```python
results_df["%_non_empty_rf"]   = results_df["non_empty_rf"]   / (results_df["non_empty_rf"]   + results_df["empty_rf"])
results_df["%_non_empty_mgmt"] = results_df["non_empty_mgmt"] / (results_df["non_empty_mgmt"] + results_df["empty_mgmt"])
```
